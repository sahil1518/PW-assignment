```python
# Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and
# y-values as the sine of x.
from bokeh.plotting import figure, show
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)


p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)')


p.line(x, y, legend_label="sin(x)", line_width=2)

show(p)



```

```python
# 2.Create a Bokeh scatter plot using randomly generated x and y values.
# Use different sizes and colors for the markers based on the 'sizes' and
# 'colors' columns.

from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
import numpy as np
import pandas as pd

np.random.seed(42)
data = {
    'x': np.random.rand(100) * 10,
    'y': np.random.rand(100) * 10,
    'sizes': np.random.randint(5, 20, 100),
    'colors': np.random.choice(['red', 'green', 'blue'], 100)
}
df = pd.DataFrame(data)


source = ColumnDataSource(df)

p = figure(title="Scatter Plot with Different Marker Sizes and Colors")

p.scatter('x', 'y', source=source, size='sizes', color=factor_cmap('colors', palette=['red', 'green', 'blue'], factors=['red', 'green', 'blue']), legend_field='colors')

p.legend.title = 'Colors'

show(p)

```

```python
 # 3.

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))


p = figure(x_range=fruits, height=350, title="Fruit Counts")
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits")


p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = 'Counts'

show(p)

```

```python
# 4.
import numpy as np
from bokeh.plotting import figure, show


data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)


p = figure(title="Histogram of Random Data", x_axis_label="Value", y_axis_label="Frequency")

p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

show(p)
```

```python
# 5.
import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import LinearColorMapper, ColorBar

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1))

color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')
show(p)
```
